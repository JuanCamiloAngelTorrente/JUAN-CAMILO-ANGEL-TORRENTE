### 6. El Impacto de las Becas
**Tarea:** Calcula el promedio académico medio para los estudiantes que tienen beca y para los que no la tienen. ¿Existe una diferencia notable?

**RESPUESTA**

```python
df.groupby('Tiene_Beca')['Promedio_Academico'].mean()
```

```python
# Análisis básico del promedio académico por tipo de beca
print("=== PROMEDIO ACADÉMICO POR ESTADO DE BECA ===")

# Promedio por grupo
promedio_con_beca = df[df['Tiene_Beca'] == 'Sí']['Promedio_Academico'].mean()
promedio_sin_beca = df[df['Tiene_Beca'] == 'No']['Promedio_Academico'].mean()

print(f"Estudiantes CON beca: {promedio_con_beca:.3f}")
print(f"Estudiantes SIN beca: {promedio_sin_beca:.3f}")
print(f"Diferencia: {abs(promedio_con_beca - promedio_sin_beca):.3f}")

# Análisis más detallado
print("\n=== ESTADÍSTICAS DETALLADAS ===")
estadisticas_beca = df.groupby('Tiene_Beca')['Promedio_Academico'].agg([
    'count', 'mean', 'median', 'std', 'min', 'max'
]).round(3)
print(estadisticas_beca)

# Visualización con boxplot
plt.figure(figsize=(10, 6))
df.boxplot(column='Promedio_Academico', by='Tiene_Beca', figsize=(8, 6))
plt.title('Distribución del Promedio Académico por Estado de Beca')
plt.xlabel('Tiene Beca')
plt.ylabel('Promedio Académico')
plt.suptitle('')  # Quitar título automático
plt.show()

# Gráfico de barras para comparar promedios
plt.figure(figsize=(8, 6))
grupos = ['Con Beca', 'Sin Beca']
promedios = [promedio_con_beca, promedio_sin_beca]
plt.bar(grupos, promedios, color=['lightblue', 'lightcoral'], alpha=0.7, edgecolor='black')
plt.ylabel('Promedio Académico')
plt.title('Comparación de Promedios Académicos por Estado de Beca')
plt.ylim(min(promedios) - 0.2, max(promedios) + 0.2)

# Añadir valores sobre las barras
for i, v in enumerate(promedios):
    plt.text(i, v + 0.02, f'{v:.3f}', ha='center', fontweight='bold')

plt.show()

# Conteo de estudiantes por grupo
print("\n=== DISTRIBUCIÓN DE LA MUESTRA ===")
conteo_beca = df['Tiene_Beca'].value_counts()
print(conteo_beca)
print(f"\nPorcentajes:")
print((conteo_beca / conteo_beca.sum() * 100).round(1))
```

### 9. ¿Afecta el Tiempo de Transporte al Rendimiento?
**Tarea:** Utiliza un diagrama de caja (`boxplot`) para comparar la distribución del `Promedio_Academico` según el `Tiempo_Transporte`. ¿Los estudiantes que viven más cerca tienen, en general, mejores promedios?

```python
plt.figure(figsize=(14, 9))
sns.boxplot(x='Tiempo_Transporte', y='Promedio_Academico', data=df, order=sorted(df['Tiempo_Transporte'].unique()))
plt.xticks(rotation=90)   # rotar etiquetas para que no se encimen
plt.tight_layout()        # evita que se corte el gráfico
plt.title('Distribución del Promedio Academico según el Tiempo de Transporte')
plt.xlabel('Tiempo_Transporte')
plt.ylabel('Promedio_Academico')
```

Son muchos valores en el eje X, por lo que se tiene que agrupar los datos en tiempo para mejorar el analisis.

```python
bins = [0, 20, 40, 60, 80, 100, 120]
labels = ["0-20", "21-40", "41-60", "61-80", "81-100", "101-119"]

# Crear nueva columna categorizada
df["Transporte_Categ"] = pd.cut(df["Tiempo_Transporte"], bins=bins, labels=labels, include_lowest=True)

# Boxplot usando categorías
plt.figure(figsize=(10,6))
sns.boxplot(x="Transporte_Categ", y="Promedio_Academico", data=df)

# Personalización
plt.title("Promedio Académico según Tiempo de Transporte (agrupado en categorías)")
plt.xlabel("Tiempo de Transporte (minutos)")
plt.ylabel("Promedio Académico")
plt.tight_layout()
plt.show()
```

```python
corr = df['Tiempo_Transporte'].corr(df['Promedio_Academico'], method='pearson')
print("Correlación de Pearson:", corr)
```

**RESPUESTA**

Los promedios acàdemicos son bastante estables en todos los rangos, encopntrandose entre 3,3 y 3,7.

No se observa  una tendencia lineal clara, el grupo de 81 a 100 minutos si muestra un rendimiento ligeramente inferior, pero no parece muy marcada.

El factor transporte no parece ser un determiante del rendimiento academico en la muestra, ademàs se observa una correlaciòn negativa, lo que indica que no existe una relaciòn.
