### Punto 6: Estandarizar la Columna 'Genero'
- Convierte todo el texto a minúsculas.
- Corrige los errores tipográficos (ej. 'masc' -> 'masculino', 'femenin' -> 'femenino').
- Muestra los valores únicos de nuevo para verificar la corrección.


```python
# Tu código aquí
# Convertir todo a minúsculas
df["Genero"] = df["Genero"].str.lower()

# Corregir errores tipográficos comunes
df["Genero"] = df["Genero"].replace({
    "masc": "masculino",
    "masculin": "masculino",
    "m": "masculino",
    "fem": "femenino",
    "femenin": "femenino",
    "f": "femenino"
})

# Verificar valores únicos después de la corrección
print("Valores únicos en 'Genero' después de estandarizar:")
print(df["Genero"].unique())

df['Genero'].value_counts()
```

### Punto 7: Estandarizar la Columna 'Suscrito'
- Unifica los valores a solo dos categorías: 'si' y 'no'.
- Muestra los valores únicos para verificar.


```python
# Tu código aquí
# Pasar todo a minúsculas y quitar espacios extra
df["Suscrito"] = df["Suscrito"].str.lower().str.strip()

# Reemplazar variaciones comunes
df["Suscrito"] = df["Suscrito"].replace({
    "si": "sí",
    "s": "sí",
    "yes": "sí",
    "y": "sí",
    "noo": "no",
    "n": "no"
})

# Verificar valores únicos después de la corrección
print("Valores únicos en 'Suscrito' después de estandarizar:")
print(df["Suscrito"].unique())
```

### Punto 9: Imputar Valores Nulos en 'Ciudad'
Utiliza `SimpleImputer` para reemplazar los valores `NaN` en la columna `Ciudad` con la **moda** (el valor más frecuente).


```python
# Tu código aquí - Se reemplazan los datos para que los valores de ciudad sean uniformes
# Ya arriba se normalizaron los datos de ciudad con la MODA
df['Ciudad'] = df['Ciudad'].str.lower() ##Pasa los datos a minuscula
df['Ciudad'] = df['Ciudad'].replace('medelin', 'medellin')
df['Ciudad'] = df['Ciudad'].replace('medellã­n', 'medellin')
df['Ciudad'] = df['Ciudad'].replace('bogotã¡', 'bogota')
df['Ciudad'] = df['Ciudad'].replace('cartgena', 'cartagena')
df['Ciudad'] = df['Ciudad'].replace('bog', 'bogota')
df['Ciudad'].value_counts()
```

### Punto 10: Aplicar Label Encoding a 'Suscrito'
Usa `LabelEncoder` para transformar la columna `Suscrito` en valores numéricos (0 y 1).


```python
# Tu código aquí
from sklearn.preprocessing import LabelEncoder

# Crear el codificador
encoder = LabelEncoder()

# Aplicar el encoding a la columna 'Suscrito'
df["Suscrito"] = encoder.fit_transform(df["Suscrito"])

# Verificar los valores transformados
print("Valores únicos en 'Suscrito' después del Label Encoding:")
print(df["Suscrito"].unique())

# Mostrar cómo quedó el mapeo
print("\nMapeo de valores:")
for original, encoded in zip(encoder.classes_, encoder.transform(encoder.classes_)):
    print(f"{original} -> {encoded}")
```

### Punto 11: Aplicar One-Hot Encoding a 'Genero'
- Utiliza `pd.get_dummies()` para crear nuevas columnas para cada categoría de 'Genero'.
- Une estas nuevas columnas al DataFrame principal y elimina la columna 'Genero' original.


```python
# Tu código aquí
# Crear variables dummies para la columna 'Genero'
dummies_genero = pd.get_dummies(df['Genero'], prefix='Genero')

# Unir las variables dummies al DataFrame principal
df = pd.concat([df, dummies_genero], axis=1)

# Eliminar la columna original 'Genero'
df = df.drop('Genero', axis=1)

# Mostrar el DataFrame actualizado
df.head()
```

### Punto 13: Normalización (Min-Max Scaling)
Aplica `MinMaxScaler` a la columna `Ingresos_Anuales` para escalar sus valores al rango [0, 1].


```python
# Tu código aquí
from sklearn.preprocessing import MinMaxScaler

# Crear el escalador
scaler = MinMaxScaler()

# Aplicar Min-Max Scaling a la columna 'Ingresos_Anuales'
df['Ingresos_Anuales_Normalizados'] = scaler.fit_transform(df[['Ingresos_Anuales']])

# Mostrar el DataFrame actualizado
df.head()
```

### Punto 14: Estandarización (Standard Scaler)
Aplica `StandardScaler` a la columna `Edad` para que tenga una media de 0 y una desviación estándar de 1.


```python
# Tu código aquí
from sklearn.preprocessing import StandardScaler

# Crear el escalador
scaler = StandardScaler()

# Aplicar StandardScaler a la columna 'Edad'
df['Edad_Estandarizada'] = scaler.fit_transform(df[['Edad']])

# Mostrar el DataFrame actualizado
df.head()
```
